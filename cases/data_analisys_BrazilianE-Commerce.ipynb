### 2.4 Análise da diferença dos valores

```python
vendas = pd.read_excel('../data/base_vendas.xlsx')
```

```python
pagamentos = pd.read_excel('../data/base_pagamentos.xlsx')
```

```python
# Agregando a base de vendas pelo preço e frete:
agg_vendas = vendas.groupby('order_id')[['price','freight_value']].sum()
```

```python
# Agregando a base de pagamentos pelo valor pago:
agg_pagamentos = pagamentos.groupby('order_id')['payment_value'].sum()
```

```python
# Concatenando essas duas bases:
agg = pd.concat([agg_vendas,agg_pagamentos],axis=1)
agg.head()
```

```python
agg.index.name = None # Retirando o índice de nome
agg
```

```python
# Criando uma coluna para calcular o valor total da compra:
agg['total'] = agg.price+agg.freight_value
```

```python
# Calculando a diferença entre os valores:
agg['Dif'] = agg.total - agg.payment_value
```

```python
# Filtrando para valores onde o valor da compra é vazio:
agg.loc[agg.price == 0,'payment_value'].sum()
```

```python
# Localizando as colunas onde existe uma diferença entre o valor da compra e o valor pago:
agg.loc[agg.price != 0,'Dif'].sum()
```

```python
# Visualizando essa tabela:
agg.loc[(abs(agg.Dif) > 10) & (agg.price != 0)]
```
